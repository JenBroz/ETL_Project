```python
# SA = co_homes['Street Address']
SA = []
for x in co_homes['Name']:
    SA.append(x)
                  
```

```python
SAS = SA[:10]
SAS
```

```python
# lat.append(response['coord']['lat'])
# lng.append(response['coord']['lon'])
Lat=[]
Lng=[]
Street_Address=[]
count=1

for address in SA:
    try:
        request = (f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={g_key}')
        response= requests.get(request).json()
        print(f'retrieving result {count} from: {address}')
        Lat.append(response['results'][0]['geometry']['location']['lat'])
        Lng.append(response['results'][0]['geometry']['location']['lng'])
        Street_Address.append(address)
        count = count + 1
    except:
        print('error retrieving coordinates skipping')
        count = count + 1
print('complete')
```

```python
co_rent_geo = pd.DataFrame(
    {'Name': Street_Address,
     'Lat': Lat,
     'Lng': Lng
    })

co_geo.head()
```

```python
co_rent_geo.to_csv(r'../Data/co_rent_geo.csv', index = False)
```
